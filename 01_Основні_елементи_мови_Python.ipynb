# –ó–º—ñ–Ω–Ω—ñ –≤ Python

| `set` (–º–Ω–æ–∂–∏–Ω–∞)              | `colors = {"red", "green"}`          | –ù–µ–≤–ø–æ—Ä—è–¥–∫–æ–≤–∞–Ω–∞ –∫–æ–ª–µ–∫—Ü—ñ—è —É–Ω—ñ–∫–∞–ª—å–Ω–∏—Ö –µ–ª–µ–º–µ–Ω—Ç—ñ–≤.                |
| `dict` (—Å–ª–æ–≤–Ω–∏–∫)             | `user = {"name": "Ivan", "age": 15}` | –ù–∞–±—ñ—Ä –ø–∞—Ä ¬´–∫–ª—é—á‚Äì–∑–Ω–∞—á–µ–Ω–Ω—è¬ª.                                   |
| `NoneType`                   | `x = None`                           | –°–ø–µ—Ü—ñ–∞–ª—å–Ω–∏–π —Ç–∏–ø, —â–æ –æ–∑–Ω–∞—á–∞—î ¬´–ø–æ—Ä–æ–∂–Ω—î –∑–Ω–∞—á–µ–Ω–Ω—è¬ª –∞–±–æ ¬´–Ω—ñ—á–æ–≥–æ¬ª. |

–£ –Ω–∞—Å—Ç—É–ø–Ω–æ–º—É –±–ª–æ—Ü—ñ –ø—Ä–∏–∫–ª–∞–¥–∏ –∑–º—ñ–Ω–Ω–∏—Ö –≤—ñ–¥–ø–æ–≤—ñ–¥–Ω–∏—Ö —Ç–∏–ø—ñ–≤:

```python
# int ‚Äì —Ü—ñ–ª—ñ —á–∏—Å–ª–∞
x = 42
print(x, type(x))  # 42 <class 'int'>

# float ‚Äì —á–∏—Å–ª–∞ –∑ –¥—Ä–æ–±–æ–≤–æ—é —á–∞—Å—Ç–∏–Ω–æ—é
pi = 3.14
print(pi, type(pi))  # 3.14 <class 'float'>

# complex ‚Äì –∫–æ–º–ø–ª–µ–∫—Å–Ω—ñ —á–∏—Å–ª–∞
z = 2 + 3j
print(z, type(z))  # (2+3j) <class 'complex'>

# str ‚Äì —Ä—è–¥–∫–∏ (—Ç–µ–∫—Å—Ç)
name = "Python"
print(name, type(name))  # Python <class 'str'>

# bool ‚Äì –ª–æ–≥—ñ—á–Ω—ñ –∑–Ω–∞—á–µ–Ω–Ω—è
flag = True
print(flag, type(flag))  # True <class 'bool'>

# list ‚Äì —É–ø–æ—Ä—è–¥–∫–æ–≤–∞–Ω–∞ –∑–º—ñ–Ω—é–≤–∞–Ω–∞ –∫–æ–ª–µ–∫—Ü—ñ—è
numbers = [1, 2, 3]
print(numbers, type(numbers))  # [1, 2, 3] <class 'list'>

# tuple ‚Äì —É–ø–æ—Ä—è–¥–∫–æ–≤–∞–Ω–∞ –Ω–µ–∑–º—ñ–Ω—é–≤–∞–Ω–∞ –∫–æ–ª–µ–∫—Ü—ñ—è
point = (10, 20)
print(point, type(point))  # (10, 20) <class 'tuple'>

# set ‚Äì –Ω–µ–≤–ø–æ—Ä—è–¥–∫–æ–≤–∞–Ω–∞ –∫–æ–ª–µ–∫—Ü—ñ—è —É–Ω—ñ–∫–∞–ª—å–Ω–∏—Ö –µ–ª–µ–º–µ–Ω—Ç—ñ–≤
colors = {"red", "green"}
print(colors, type(colors))  # {'red', 'green'} <class 'set'>

# dict ‚Äì —Å–ª–æ–≤–Ω–∏–∫, –Ω–∞–±—ñ—Ä –ø–∞—Ä "–∫–ª—é—á‚Äì–∑–Ω–∞—á–µ–Ω–Ω—è"
user = {"name": "Ivan", "age": 15}
print(user, type(user))  # {'name': 'Ivan', 'age': 15} <class 'dict'>

# NoneType ‚Äì —Å–ø–µ—Ü—ñ–∞–ª—å–Ω–∏–π —Ç–∏–ø "–Ω—ñ—á–æ–≥–æ"
x = None
print(x, type(x))  # None <class 'NoneType'>
```

–¢–∏–ø –∑–º—ñ–Ω–Ω–æ—ó –º–æ–∂–µ –Ω–µ —Ç—ñ–ª—å–∫–∏ –¥–∏–Ω–∞–º—ñ—á–Ω–æ –≤–∏–∑–Ω–∞—á–∞—Ç–∏—Å—è, –∞–ª–µ –π –¥–∏–Ω–∞–º—ñ—á–Ω–æ –∑–º—ñ–Ω—é–≤–∞—Ç–∏—Å—è. –ù–∞–ø—Ä–∏–∫–ª–∞–¥:

```python
x = 42
print(x, type(x))
x = "Hello Python"
print(x, type(x))
```

## –ü—Ä–∏—Å–≤–æ—î–Ω–Ω—è –∑–Ω–∞—á–µ–Ω—å –∑–º—ñ–Ω–Ω–∏–º

–î–ª—è –ø—Ä–∏—Å–≤–æ—î–Ω–Ω—è –∑–Ω–∞—á–µ–Ω–Ω—è –∑–º—ñ–Ω–Ω—ñ–π (–ø—Ä–æ—Å—Ç—ñ—à–µ: –¥–ª—è —Ç–æ–≥–æ, —â–æ–± –∑–∞–ø–∏—Å–∞—Ç–∏ –∑–Ω–∞—á–µ–Ω–Ω—è —É –∑–º—ñ–Ω–Ω—É) —É Python –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—î—Ç—å—Å—è –æ–ø–µ—Ä–∞—Ç–æ—Ä `=`:

```python
some_var = 56 # –£ –∑–º—ñ–Ω–Ω—É some_var –∑–∞–ø–∏—Å–∞–ª–∏ –∑–Ω–∞—á–µ–Ω–Ω—è 56
```

–ö—Ä—ñ–º —Ç–æ–≥–æ, Python –¥–æ–∑–≤–æ–ª—è—î –≤ –æ–¥–Ω–æ–º—É —Ä—è–¥–∫—É –ø—Ä–∏—Å–≤–æ—é–≤–∞—Ç–∏ –∑–Ω–∞—á–µ–Ω–Ω—è –æ–¥—Ä–∞–∑—É –∫—ñ–ª—å–∫–æ–º –∑–º—ñ–Ω–Ω–∏–º:

```python
var_1 = var_2 = var_3 = 456
print(f"{var_1}, {var_2}, {var_3}")
```

–¢–∞–∫–æ–∂ –º–æ–∂–Ω–∞ –≤ –æ–¥–Ω–æ–º—É —Ä—è–¥–æ—á–∫—É –ø—Ä–æ—Å–≤–æ—é–≤–∞—Ç–∏ —Ä—ñ–∑–Ω—ñ –∑–Ω–∞—á–µ–Ω–Ω—è —Ä—ñ–∑–Ω–∏–º –∑–º—ñ–Ω–Ω–∏–º:

```python
v1, v2, v3 = 23, 34, 56
print(f"{v1}, {v2}, {v3}")
```

–ö—ñ–ª—å–∫—ñ—Å—Ç—å –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ –ª—ñ–≤–æ—Ä—É—á —ñ –ø—Ä–∞–≤–æ—Ä—É—á –≤—ñ–¥ –æ–ø–µ—Ä–∞—Ç–æ—Ä–∞ –ø—Ä–∏—Å–≤–æ—é–≤–∞–Ω–Ω—è –º–∞—î –±—É—Ç–∏ –æ–¥–Ω–∞–∫–æ–≤–∞, —ñ–Ω–∞–∫—à–µ –±—É–¥–µ –≤–∏–¥–∞–Ω–æ –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è –ø—Ä–æ —Å–∏–Ω—Ç–∞–∫—Å–∏—á–Ω—É –ø–æ–º–∏–ª–∫—É

```python
v1, v2, v3 = 23, 34, 56, 37
print(f"{v1}, {v2}, {v3}")
```

–î–ª—è —Ç–æ–≥–æ, —â–æ–± —É–Ω–∏–∫–Ω—É—Ç–∏ —Ü—ñ—î—ó –ø–æ–º–∏–ª–∫–∏ –º–æ–∂–Ω–∞ –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–≤–∞—Ç–∏ –æ–ø–µ—Ä–∞—Ç–æ—Ä `*`. –ô–æ–≥–æ —Ç—Ä–µ–±–∞ –ø–æ—Å—Ç–∞–≤–∏—Ç–∏ –ø–µ—Ä–µ–¥ —ñ–¥–µ–Ω—Ç–∏—Ñ—ñ–∫–∞—Ç–æ—Ä–æ–º –æ–¥–Ω—ñ—î—ó –∑—ñ –∑–º—ñ–Ω–Ω–∏—Ö, —Ç–æ–¥—ñ –≤ —Ü—é –∑–º—ñ–Ω–Ω—É –∑–∞–ø–∏—à–µ—Ç—å—Å—è —Å–ø–∏—Å–æ–∫ –∑–Ω–∞—á–µ–Ω—å, —è–∫—ñ –Ω–µ –≤–ª—ñ–∑–∞—é—Ç—å –≤ —ñ–Ω—à—ñ –∑–º—ñ–Ω–Ω—ñ:

```python
v1, *v2, v3 = 23, 34, 56, 37
print(f"{v1}, {v2}, {v3}")
```

```python
v1, v2, *v3 = 23, 34, 56, 37
print(f"{v1}, {v2}, {v3}")
```
